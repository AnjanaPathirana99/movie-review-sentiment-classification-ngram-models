# 2) Text pre-processing

```python
import re
import string
```

## Lowercasing

```python
positive_corpus = [positive_review.lower() for positive_review in positive_reviews]
positive_corpus
```

```python
negative_corpus = [negative_review.lower() for negative_review in negative_reviews]
negative_corpus
```

## Remove special characters (, . " ! ')

```python
# Remove special characters from the positive corpus while only keeping alphanumeric characters and spaces
def remove_special_characters(positive_corpus):

    cleaned_positive_corpus = [''.join(char for char in sentence if char.isalnum() or char.isspace()) for sentence in positive_corpus]

    return cleaned_positive_corpus

cleaned_positive_corpus = remove_special_characters(positive_corpus)
cleaned_positive_corpus
```

```python
# Remove special characters from the negative corpus while only keeping alphanumeric characters and spaces
def remove_special_characters(negative_corpus):

    cleaned_negative_corpus = [''.join(char for char in sentence if char.isalnum() or char.isspace()) for sentence in negative_corpus]

    return cleaned_negative_corpus

cleaned_negative_corpus = remove_special_characters(negative_corpus)
cleaned_negative_corpus
```

## Tokenizing

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
# Tokenize cleaned_positive_corpus
tokenized_cleaned_positive_corpus = [word_tokenize(review) for review in cleaned_positive_corpus]
tokenized_cleaned_positive_corpus
```

```python
# Tokenize cleaned_negative_corpus
tokenized_cleaned_negative_corpus = [word_tokenize(review) for review in cleaned_negative_corpus]
tokenized_cleaned_negative_corpus
```

## Remove stop words

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

import nltk
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))

filtered_tokenized_positive_corpus = [
    [word for word in review if word.lower() not in stop_words]
    for review in tokenized_cleaned_positive_corpus
]

filtered_tokenized_positive_corpus
```

```python
stop_words = set(stopwords.words('english'))

filtered_tokenized_negative_corpus = [
    [word for word in review if word.lower() not in stop_words]
    for review in tokenized_cleaned_negative_corpus
]

filtered_tokenized_negative_corpus
```

## Stemming

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

porter = PorterStemmer()
```

```python
cleaned_positive_corpus = [
    [porter.stem(word) for word in review]
    for review in filtered_tokenized_positive_corpus
]

cleaned_positive_corpus
```

```python
cleaned_negative_corpus = [
    [porter.stem(word) for word in review]
    for review in filtered_tokenized_negative_corpus
]

cleaned_negative_corpus
```

# 3) N Gram Models (Best choise for N is explained last)

```python
from nltk import ngrams
```

```python
# N Gram models for tokenized_cleaned_positive_corpus

# Function to add <s> and </s> tags to each sentence
def add_start_end_tags(sentences, n):
    start_tag = ["<s>"] * (n - 1)
    end_tag = ["</s>"] * (n - 1)
    return [start_tag + sentence + end_tag for sentence in sentences]

# Unigrams per sentence
unigrams_per_sentence_positive_corpus = [[word for word in sentence] for sentence in cleaned_positive_corpus]

# Bigrams per sentence (with <s> and </s> tags)
bigrams_per_sentence_positive_corpus = [list(ngrams(sentence, 2)) for sentence in add_start_end_tags(unigrams_per_sentence_positive_corpus, 2)]

# Trigrams per sentence (with <s><s> and </s></s> tags)
trigrams_per_sentence_positive_corpus = [list(ngrams(sentence, 3)) for sentence in add_start_end_tags(unigrams_per_sentence_positive_corpus, 3)]

# Print the results
print("Unigrams per sentence:")
for sentence in unigrams_per_sentence_positive_corpus:
    print(sentence)

print("\nBigrams per sentence (with <s> and </s> tags):")
for sentence in bigrams_per_sentence_positive_corpus:
    print(sentence)

print("\nTrigrams per sentence (with <s><s> and </s></s> tags):")
for sentence in trigrams_per_sentence_positive_corpus:
    print(sentence)
```

```python
# N Gram models for tokenized_cleaned_negative_corpus

# Function to add <s> and </s> tags to each sentence
def add_start_end_tags(sentences, n):
    start_tag = ["<s>"] * (n - 1)
    end_tag = ["</s>"] * (n - 1)
    return [start_tag + sentence + end_tag for sentence in sentences]

# Unigrams per sentence
unigrams_per_sentence_negative_corpus = [[word for word in sentence] for sentence in cleaned_negative_corpus]

# Bigrams per sentence (with <s> and </s> tags)
bigrams_per_sentence_negative_corpus = [list(ngrams(sentence, 2)) for sentence in add_start_end_tags(unigrams_per_sentence_negative_corpus, 2)]

# Trigrams per sentence (with <s><s> and </s></s> tags)
trigrams_per_sentence_negative_corpus = [list(ngrams(sentence, 3)) for sentence in add_start_end_tags(unigrams_per_sentence_negative_corpus, 3)]

print("Unigrams per sentence:")
for sentence in unigrams_per_sentence_negative_corpus:
    print(sentence)

print("\nBigrams per sentence:")
for sentence in bigrams_per_sentence_negative_corpus:
    print(sentence)

print("\nTrigrams per sentence:")
for sentence in trigrams_per_sentence_negative_corpus:
    print(sentence)
```

# 4) Calculate N Gram probabilities

```python
from nltk import ngrams
from collections import Counter
```

## Positive corpus

```python
# Assuming you have the cleaned_positive_corpus as provided

# Flatten the list of sentences into a list of words
flat_corpus = [word for sentence in cleaned_positive_corpus for word in sentence]

# Unigrams
unigram_counts = Counter(flat_corpus)
total_unigrams = sum(unigram_counts.values())
unigram_postive_probabilities = {unigram: count / total_unigrams for unigram, count in unigram_counts.items()}

# Bigrams
bigram_counts = Counter(ngrams(flat_corpus, 2))
total_bigrams = sum(bigram_counts.values())
bigram_positive_probabilities = {bigram: count / total_bigrams for bigram, count in bigram_counts.items()}

# Trigrams
trigram_counts = Counter(ngrams(flat_corpus, 3))
total_trigrams = sum(trigram_counts.values())
trigram_positive_probabilities = {trigram: count / total_trigrams for trigram, count in trigram_counts.items()}

# Print the results
print("Unigram Probabilities:")
print(unigram_postive_probabilities)

print("\nBigram Probabilities:")
print(bigram_positive_probabilities)

print("\nTrigram Probabilities:")
print(trigram_positive_probabilities)
```

## Compare the N-gram probability of test_corpus with respect to positive & negative corpus

```python
import math

# Calculate log probability for each N-gram in the test_corpus
# Consideer log due to multiplication of very small numbers
positive_log_probability_unigram = sum([math.log(unigram_postive_probabilities.get(gram, 1e-10)) for gram in test_corpus_unigram_probabilities])
positive_log_probability_bigram = sum([math.log(bigram_positive_probabilities.get(gram, 1e-10)) for gram in test_corpus_bigram_probabilities])
positive_log_probability_trigram = sum([math.log(trigram_positive_probabilities.get(gram, 1e-10)) for gram in test_corpus_trigram_probabilities])

negative_log_probability_unigram = sum([math.log(unigram_negative_probabilities.get(gram, 1e-10)) for gram in test_corpus_unigram_probabilities])
negative_log_probability_bigram = sum([math.log(bigram_negative_probabilities.get(gram, 1e-10)) for gram in test_corpus_bigram_probabilities])
negative_log_probability_trigram = sum([math.log(trigram_negative_probabilities.get(gram, 1e-10)) for gram in test_corpus_trigram_probabilities])

# Take exponential to get antilog
positive_probability_unigram = math.exp(positive_log_probability_unigram)
positive_probability_bigram = math.exp(positive_log_probability_bigram)
positive_probability_trigram = math.exp(positive_log_probability_trigram)

negative_probability_unigram = math.exp(negative_log_probability_unigram)
negative_probability_bigram = math.exp(negative_log_probability_bigram)
negative_probability_trigram = math.exp(negative_log_probability_trigram)

print("Probability (Positive unigram):", positive_probability_unigram)
print("Probability (Positive bigram):", positive_probability_bigram)
print("Probability (Positive trigram):", positive_probability_trigram)

print("Probability (Negative unigram):", negative_probability_unigram)
print("Probability (Negative bigram):", negative_probability_bigram)
print("Probability (Negative trigram):", negative_probability_trigram)

# sum_positive_probability = positive_probability_unigram + positive_probability_bigram + positive_probability_trigram
# sum_negative_probability = negative_probability_unigram + negative_probability_bigram + negative_probability_trigram

# print("Probability (Positive sum):", sum_positive_probability)
# print("Probability (Negative sum):", sum_negative_probability)
```
